Q2. Write a NumPy program to generate a 2D array and performs a discrete Fourier transform
using SciPy's fftpack module.

```python
import numpy as np
from scipy import fftpack

# Generate a 2D NumPy array (e.g., 4x4)
array_2d = np.random.rand(4, 4)

# Perform 2D Discrete Fourier Transform
dft_2d = fftpack.fft2(array_2d)

# Shift zero-frequency component to center
dft_shifted = fftpack.fftshift(dft_2d)

# Display results
print("Original 2D Array:")
print(array_2d)

print("\n2D Discrete Fourier Transform:")
print(dft_2d)

print("\nShifted DFT (Centered Frequency):")
print(dft_shifted)
```

Q3. Write a NumPy program that creates a NumPy array and uses SciPy to solve a linear
algebra problem. For example, finding the determinant, inverse, and eigenvalues of a matrix

```python
import numpy as np
from scipy import linalg

# Create a NumPy array (square matrix)
A = np.array([[4, 2],
              [1, 3]])

# Compute determinant
det_A = linalg.det(A)

# Compute inverse
inv_A = linalg.inv(A)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = linalg.eig(A)

# Display results
print("Matrix A:")
print(A)

print("\nDeterminant of A:")
print(det_A)

print("\nInverse of A:")
print(inv_A)

print("\nEigenvalues of A:")
print(eigenvalues)

print("\nEigenvectors of A:")
print(eigenvectors)
```

Q4. Write a NumPy program to generate a set of data points using NumPy and perform
interpolation using SciPy's interpolate module.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

# Generate data points
x = np.linspace(0, 10, 10)
y = np.sin(x)

# Create interpolation function
interp_func = interpolate.interp1d(x, y, kind='cubic')

# Generate new x values for interpolation
x_new = np.linspace(0, 10, 100)
y_new = interp_func(x_new)

# Display results
print("Original x values:", x)
print("Original y values:", y)

# Plot original points and interpolated curve
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Interpolated Curve')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolation using SciPy")
plt.show()
```

Q7. This project involves analysing the marks of students across multiple subjects. The goal is
to generate and analyse data, including calculating total marks, average marks, subject-wise

percentage and generate insights based on students' performance.
Student Name Math Physics Chemistry English
Arin 85 78 92 88
Aditya 79 82 74 90
Chirag 90 85 89 92
Gurleen 66 75 80 78
Kunal 70 68 75 85


```python
import numpy as np

# Student names and subjects
students = np.array(["Arin", "Aditya", "Chirag", "Gurleen", "Kunal"])
subjects = np.array(["Math", "Physics", "Chemistry", "English"])

# Marks data (rows = students, columns = subjects)
marks = np.array([
    [85, 78, 92, 88],
    [79, 82, 74, 90],
    [90, 85, 89, 92],
    [66, 75, 80, 78],
    [70, 68, 75, 85]
])

print("Marks Matrix:\n", marks)

# Total and average marks per student
total_marks = np.sum(marks, axis=1)
average_marks = np.mean(marks, axis=1)

# Subject-wise average performance
subject_average = np.mean(marks, axis=0)

# Top and bottom performers
top_student_index = np.argmax(total_marks)
bottom_student_index = np.argmin(total_marks)

# Passing criteria (assume pass mark = 40)
pass_mark = 40
passed_students = np.all(marks >= pass_mark, axis=1)
passing_percentage = (np.sum(passed_students) / len(students)) * 100

# Display results
print("\nTotal and Average Marks:")
for i in range(len(students)):
    print(f"{students[i]} - Total: {total_marks[i]}, Average: {average_marks[i]:.2f}")

print("\nSubject-wise Average Marks:")
for i in range(len(subjects)):
    print(f"{subjects[i]}: {subject_average[i]:.2f}")

print("\nTop Performer:", students[top_student_index],
      "with Total Marks =", total_marks[top_student_index])

print("Bottom Performer:", students[bottom_student_index],
      "with Total Marks =", total_marks[bottom_student_index])

print("\nPassing Percentage:", passing_percentage, "%")
```

Q8. You are provided with the following data for the velocity of an object over time. The
velocity data is modelled by a quadratic function of the form v(t)=at2+bt+c , where v(t) is the
velocity at time t, and a, b, and c are constants to be determined using curve fitting. (Using
SciPy's Curve fitting)
Time (Seconds) 0 1 2 3 4 5
Velocity (m/s) 2 3.1 7.9 18.2 34.3 56.2

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Given data
time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

# Define quadratic model v(t) = a*t^2 + b*t + c
def velocity_model(t, a, b, c):
    return a*t**2 + b*t + c

# Perform curve fitting
params, covariance = curve_fit(velocity_model, time, velocity)

a, b, c = params

# Display results
print("Fitted Parameters:")
print("a =", a)
print("b =", b)
print("c =", c)

# Generate fitted curve
time_fit = np.linspace(0, 5, 100)
velocity_fit = velocity_model(time_fit, a, b, c)

# Plot results
plt.scatter(time, velocity, label="Given Data", color="red")
plt.plot(time_fit, velocity_fit, label="Fitted Curve")
plt.xlabel("Time (seconds)")
plt.ylabel("Velocity (m/s)")
plt.title("Quadratic Curve Fitting of Velocity-Time Data")
plt.legend()
plt.show()
```

Q11. You are given the following data representing the population of a town in various years,
Table 1
S. No. Year Population (in thousands)
1 2000 50
2 2005 55
3 2010 70
4 2015 80
5 2020 90
What is the Pearson's correlation coefficient for the above data? Estimate the population of
the town in the year 2008 using linear interpolation/regression equation based on Table 1
data. Write Python code to perform the interpolation using Scipy functions and plot it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate, stats

# Given data
years = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([50, 55, 70, 80, 90])  # in thousands

# 1. Pearson's Correlation Coefficient
correlation, _ = stats.pearsonr(years, population)
print("Pearson's Correlation Coefficient:", correlation)

# 2. Linear interpolation using SciPy
interp_func = interpolate.interp1d(years, population, kind='linear')

# Estimate population in year 2008
year_estimate = 2008
population_2008 = interp_func(year_estimate)
print("Estimated Population in 2008:", population_2008, "thousands")

# Generate smooth values for plotting
years_new = np.linspace(2000, 2020, 100)
population_new = interp_func(years_new)

# 3. Plot original data and interpolation
plt.figure(figsize=(8, 5))
plt.scatter(years, population, color='red', label='Given Data')
plt.plot(years_new, population_new, color='blue', label='Linear Interpolation')
plt.scatter(year_estimate, population_2008, color='green',
            label='Estimated Population (2008)', zorder=5)

# Plot features
plt.xlabel("Year")
plt.ylabel("Population (in thousands)")
plt.title("Population Growth Analysis using Linear Interpolation")
plt.legend()
plt.grid(True)

plt.show()
```

Q12. Consider the polynomial equation p(x)=3x35x2+2x8
1. Use SciPy to find the roots of the polynomial.
2. Plot the polynomial function p(x) for the range of x from -3 to 3 and mark the roots on
the plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

# Define polynomial coefficients: 3x^3 - 5x^2 + 2x - 8
coefficients = [3, -5, 2, -8]

# 1. Find roots of the polynomial
roots = np.roots(coefficients)

print("Roots of the polynomial:")
print(roots)

# 2. Define the polynomial function
def p(x):
    return 3*x**3 - 5*x**2 + 2*x - 8

# Generate x values for plotting
x = np.linspace(-3, 3, 400)
y = p(x)

# Plot the polynomial
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="p(x) = 3x - 5x + 2x - 8")

# Mark the roots
for r in roots:
    if np.isreal(r):
        plt.scatter(np.real(r), 0, color='red', zorder=5)
        plt.text(np.real(r), 0, f"{np.real(r):.2f}", ha='center', va='bottom')

# Plot features
plt.axhline(0)  # x-axis
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Polynomial Function and Its Roots")
plt.legend()
plt.grid(True)

plt.show()
```

Q13. Compare the performance (time taken) of Python programs.
1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.
2. Generate a file with random text of all the MBs and convert it into upper case and
check time and plot.

```python
import time
import os
import random
import string
import matplotlib.pyplot as plt

# File sizes in MB
file_sizes = [200, 400, 600, 800, 1000]
times_taken = []

# Function to generate random text file
def generate_text_file(filename, size_mb):
    chars = string.ascii_lowercase + string.ascii_uppercase + " "
    size_bytes = size_mb * 1024 * 1024

    with open(filename, 'w') as f:
        written = 0
        while written < size_bytes:
            text = ''.join(random.choices(chars, k=1024))
            f.write(text)
            written += len(text)

# Function to convert file to uppercase
def convert_to_uppercase(input_file, output_file):
    with open(input_file, 'r') as fin, open(output_file, 'w') as fout:
        for line in fin:
            fout.write(line.upper())

# Performance testing
for size in file_sizes:
    input_file = f"random_{size}MB.txt"
    output_file = f"upper_{size}MB.txt"

    print(f"\nProcessing {size} MB file...")

    # Generate file
    generate_text_file(input_file, size)

    # Measure time
    start_time = time.time()
    convert_to_uppercase(input_file, output_file)
    end_time = time.time()

    elapsed_time = end_time - start_time
    times_taken.append(elapsed_time)

    print(f"Time taken: {elapsed_time:.2f} seconds")

    # Optional: clean up files to save space
    os.remove(input_file)
    os.remove(output_file)

# Plot performance graph
plt.figure(figsize=(8, 5))
plt.plot(file_sizes, times_taken, marker='o')
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (seconds)")
plt.title("Performance Analysis: File Size vs Uppercase Conversion Time")
plt.grid(True)
plt.show()
```

Q14. Consider the function f(x) = x
43x3+2.
1. Use SciPy to find the local minima of the function.
2. Plot the function f(x) over the range x = [2,3] and mark the local minima on the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Define the function
def f(x):
    return x**4 - 3*x**3 + 2

# 1. Find local minimum using SciPy
result = minimize_scalar(f, bounds=(-2, 3), method='bounded')

x_min = result.x
y_min = f(x_min)

print("Local minimum found at:")
print("x =", x_min)
print("f(x) =", y_min)

# 2. Plot the function
x = np.linspace(-2, 3, 400)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = x  3x + 2")
plt.scatter(x_min, y_min, color='red', zorder=5, label="Local Minimum")

# Plot features
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Function Plot with Local Minimum")
plt.legend()
plt.grid(True)

plt.show()
```

Q15. A robotic arm joint is modeled as a damped oscillatory system. The displacement of
the joint angle (t) over time can be represented by the second-order differential equation:
with initial conditions:
 (0) = 1 rad
 (0) = 0 rad/s
Tasks
1. Use SciPys odeint to solve the system of equations for t = 0  20 sec.
2. Use NumPy arrays to handle the time vector and solutions.
3. Plot the solution (t) using Matplotlib.
4. Find the maximum displacement and the time at which it occurs using NumPy
functions.
5. Comment: What does the damping coefficient 0.2 imply for robotic joint motion?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
zeta = 0.2      # damping coefficient
omega_n = 1.0   # natural frequency (rad/s)

# Define the system of ODEs
# y = [theta, theta_dot]
def damped_system(y, t):
    theta, theta_dot = y
    dtheta_dt = theta_dot
    dtheta_dot_dt = -2 * zeta * omega_n * theta_dot - omega_n**2 * theta
    return [dtheta_dt, dtheta_dot_dt]

# Time vector
t = np.linspace(0, 20, 1000)

# Initial conditions
y0 = [1, 0]  # theta(0)=1, theta'(0)=0

# Solve ODE
solution = odeint(damped_system, y0, t)

# Extract theta(t)
theta = solution[:, 0]

# Plot theta(t)
plt.figure(figsize=(8, 5))
plt.plot(t, theta, label=r'$\theta(t)$')
plt.xlabel("Time (s)")
plt.ylabel("Displacement (t) [rad]")
plt.title("Damped Oscillatory Motion of Robotic Arm Joint")
plt.grid(True)
plt.legend()
plt.show()

# Find maximum displacement and time
theta_max = np.max(theta)
t_max = t[np.argmax(theta)]
print(f"Maximum displacement _max = {theta_max:.3f} rad at t = {t_max:.3f} s")
```
