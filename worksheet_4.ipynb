Q10

```python

A = np.array([[2, 1, 3],
              [0, 5, 6],
              [7, 8, 9]])
print("10.1 Matrix A:\n", A)


print("\nDeterminant of A:", np.linalg.det(A))
print("Inverse of A:\n", np.linalg.inv(A))
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

Q11

```python

positions = np.array([(0,0), (2,3), (4,7), (7,10), (10,15)])


dist = np.linalg.norm(positions[-1] - positions[0])
print("11.1 Euclidean distance (first to last):", dist)


step_distances = np.linalg.norm(np.diff(positions, axis=0), axis=1)
total_distance = np.sum(step_distances)
print("11.2 Total distance traveled:", total_distance)
```
