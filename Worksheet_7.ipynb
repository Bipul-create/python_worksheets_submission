
Project 1: Lab Task: Tic-Tac-Toe Game in Python
Objective:
The goal of this lab task is to develop a Tic-Tac-Toe game in Python, reinforcing the students'
understanding of:
 Control flow (if-else statements, loops)

Task Details:
You will create a Tic-Tac-Toe game where two players (Player X and Player O) take turns to mark a 3x3
grid with their respective symbols ("X" or "O"). The game ends when one player wins by aligning their
symbols in a row, column, or diagonal, or when the game results in a tie.
Instructions:
1. Create the Tic-Tac-Toe board:
o Use a list of 9 elements (initialized with empty spaces) to represent the 3x3 grid.
o Write a function print_board() to display the current state of the board.
2. Handle Player Input:
o Write a function that asks the current player to choose a position (1-9) on the board.
o Validate that the input is a number between 1 and 9 and that the chosen position is
not already taken.
3. Game Logic:
o Implement a function to check if a player has won. A player wins if they have filled
any one of the following combinations with their symbol:
 Rows: [0, 1, 2], [3, 4, 5], [6, 7, 8]
 Columns: [0, 3, 6], [1, 4, 7], [2, 5, 8]
 Diagonals: [0, 4, 8], [2, 4, 6]
o Implement a function to check if the game has resulted in a tie (i.e., all positions are
filled, and no player has won).
4. Switch Turns:
o Write logic to alternate between the two players ("X" and "O") after each valid move.
5. Determine the End of the Game:
o After each move, check if a player has won or if the game ends in a tie.
o If a player wins, display a congratulatory message.
o If it’s a tie, display a message indicating the draw.
6. Replay Option:
o At the end of the game, prompt the players with the option to play again or exit.
Guidelines:
1. You must write the following functions:
o print_board(board): Prints the Tic-Tac-Toe board.
o check_winner(board, player): Checks if the current player has won the game.
o check_tie(board): Checks if the game ends in a tie.
o get_player_input(player, board): Prompts the current player for a valid move
o play_game(): The main game loop that controls the flow of the game.
2. Validate all user inputs, ensuring they enter correct positions and avoid overwriting
previously marked spots.

```python
def print_board(board):
    """
    Prints the Tic-Tac-Toe board.
    """
    print("\n")
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("---|---|---")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("---|---|---")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print("\n")


def check_winner(board, player):
    """
    Checks if the current player has won.
    """
    # Winning combinations
    win_combinations = [
        [0,1,2], [3,4,5], [6,7,8],   # Rows
        [0,3,6], [1,4,7], [2,5,8],   # Columns
        [0,4,8], [2,4,6]             # Diagonals
    ]
    for combo in win_combinations:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] == player:
            return True
    return False


def check_tie(board):
    """
    Checks if the game is a tie.
    """
    return " " not in board


def get_player_input(player, board):
    """
    Prompts the current player for a valid move.
    """
    while True:
        try:
            pos = int(input(f"Player {player}, enter your move (1-9): ")) - 1
            if pos < 0 or pos > 8:
                print("Invalid input! Enter a number between 1 and 9.")
            elif board[pos] != " ":
                print("Position already taken! Choose another spot.")
            else:
                return pos
        except ValueError:
            print("Invalid input! Enter a valid number.")


def play_game():
    """
    Main game loop controlling the Tic-Tac-Toe game.
    """
    board = [" "] * 9
    current_player = "X"
    game_over = False

    print("Welcome to Tic-Tac-Toe!")
    print_board(board)

    while not game_over:
        # Get valid player input
        move = get_player_input(current_player, board)
        board[move] = current_player

        # Print board after move
        print_board(board)

        # Check for winner
        if check_winner(board, current_player):
            print(f"Congratulations! Player {current_player} wins!")
            game_over = True
        # Check for tie
        elif check_tie(board):
            print("It's a tie!")
            game_over = True
        else:
            # Switch player
            current_player = "O" if current_player == "X" else "X"

    # Replay option
    replay = input("Do you want to play again? (yes/no): ").lower()
    if replay == "yes":
        play_game()
    else:
        print("Thank you for playing Tic-Tac-Toe!")


# Start the game
play_game()
```

Objective: The goal of this project is to create a simple command-line To-Do list application
using Python. This project will allow users to add, view, and delete tasks from their to-do list.
By completing this project, students will practice working with lists, functions, and loops.
Instructions
1. Project Overview:
You are required to develop a Python program that acts as a basic to-do list system. The
program should continuously prompt the user to either add new tasks, view current tasks, or
delete existing tasks until the user decides to exit the application.
2. Functionalities to Implement:
1. Add Task (add_task):
o Function to add a new task to the to-do list.
o Each task will be a string, and it will be stored in a list.
2. View Tasks (view_tasks):
o Display the current tasks in the to-do list, showing the task along with its
index.
3. Delete Task (delete_task):
o Function to delete a task from the list based on its index.
o If the user provides an invalid index, print an error message.
4. Exit the Program:
o A command to allow users to exit the to-do list application

```python
# Initialize an empty to-do list
todo_list = []

def add_task():
    """
    Function to add a new task to the to-do list
    """
    task = input("Enter the task you want to add: ")
    todo_list.append(task)
    print(f"Task '{task}' added successfully!")

def view_tasks():
    """
    Function to display current tasks with their indices
    """
    if not todo_list:
        print("Your to-do list is empty.")
    else:
        print("\nYour To-Do List:")
        for index, task in enumerate(todo_list, start=1):
            print(f"{index}. {task}")
        print()  # New line for readability

def delete_task():
    """
    Function to delete a task by its index
    """
    if not todo_list:
        print("Your to-do list is empty. Nothing to delete.")
        return

    try:
        index = int(input("Enter the task number to delete: "))
        if 1 <= index <= len(todo_list):
            removed_task = todo_list.pop(index - 1)
            print(f"Task '{removed_task}' deleted successfully!")
        else:
            print("Invalid task number! Please enter a valid index.")
    except ValueError:
        print("Invalid input! Please enter a numeric value.")

def main():
    """
    Main loop of the To-Do list application
    """
    print("Welcome to the Python To-Do List Application!")

    while True:
        print("\nChoose an option:")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Delete Task")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            delete_task()
        elif choice == "4":
            print("Exiting the To-Do List Application. Goodbye!")
            break
        else:
            print("Invalid choice! Please select a valid option (1-4).")

# Start the application
if __name__ == "__main__":
    main()
```

Problem Statement:
You are tasked with developing a Python-based command-line application that simulates a
robot navigating through a grid-based environment. The robot should find the shortest path
from a specified starting point to a destination while avoiding obstacles placed within the
environment. You will implement the A* pathfinding algorithm to compute the shortest path
and visualize the result.
Instructions:
1. Environment Representation:
o Use a 2D NumPy array to represent the environment (grid).
o Free spaces will be represented by 0s and obstacles by 1s.
o The user will input the grid size (rows and columns) and the number of
obstacles with their locations.
2. User Input:
o Prompt the user to input the following:
 Grid size (number of rows and columns).
 Number of obstacles and their positions in the grid.
 Starting point (coordinates) for the robot.
 Destination point (coordinates) for the robot.
3. A Pathfinding Algorithm*:
o Implement the A* algorithm to find the shortest path from the start to the
destination.
o The algorithm should calculate both:
 g: The actual distance traveled from the start point to the current point.
 h: The heuristic (Manhattan distance) from the current point to the
destination.
 f = g + h: The total cost.
o If no path is found, the program should display a message indicating that there
is no valid path.
4. Visualization:
o After calculating the path, visualize the grid using matplotlib:
 Show the robot’s path from the starting point to the destination.
 Mark the obstacles, starting point, and destination clearly on the grid.
 Obstacles should be shown as black cells, the robot's path as red dots,
the starting point as green, and the destination as blue.
5. Data Management:
o Use Pandas to display and manage the grid data before and after pathfinding.
6. Bonus:
o Provide the option for the user to retry with different start, destination, and
obstacle positions without restarting the program.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from heapq import heappush, heappop

# -------------------
# Heuristic Function
# -------------------
def manhattan_distance(a, b):
    """Calculate Manhattan distance between two points a and b"""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

# -------------------
# A* Pathfinding Algorithm
# -------------------
def astar(grid, start, end):
    rows, cols = grid.shape
    open_set = []
    heappush(open_set, (0 + manhattan_distance(start, end), 0, start, [start]))  # (f, g, current, path)
    visited = set()

    while open_set:
        f, g, current, path = heappop(open_set)

        if current in visited:
            continue
        visited.add(current)

        if current == end:
            return path  # Path found

        # Explore neighbors (up, down, left, right)
        neighbors = [
            (current[0] + 1, current[1]),
            (current[0] - 1, current[1]),
            (current[0], current[1] + 1),
            (current[0], current[1] - 1)
        ]

        for nr, nc in neighbors:
            if 0 <= nr < rows and 0 <= nc < cols and grid[nr, nc] == 0:
                if (nr, nc) not in visited:
                    g_new = g + 1
                    f_new = g_new + manhattan_distance((nr, nc), end)
                    heappush(open_set, (f_new, g_new, (nr, nc), path + [(nr, nc)]))

    return None  # No path found

# -------------------
# Visualization
# -------------------
def visualize_grid(grid, path, start, end):
    rows, cols = grid.shape
    plt.figure(figsize=(cols, rows))

    # Draw grid
    for r in range(rows):
        for c in range(cols):
            if grid[r, c] == 1:
                plt.scatter(c, rows - r - 1, color='black', s=100)  # obstacle

    # Draw path
    if path:
        path_x = [c for r, c in path]
        path_y = [rows - r - 1 for r, c in path]
        plt.plot(path_x, path_y, 'ro-', label='Path', markersize=8)

    # Draw start and end
    plt.scatter(start[1], rows - start[0] - 1, color='green', s=100, label='Start')
    plt.scatter(end[1], rows - end[0] - 1, color='blue', s=100, label='Destination')

    plt.xlim(-0.5, cols - 0.5)
    plt.ylim(-0.5, rows - 0.5)
    plt.gca().set_aspect('equal')
    plt.xticks(range(cols))
    plt.yticks(range(rows))
    plt.grid(True)
    plt.legend()
    plt.show()

# -------------------
# Main Program
# -------------------
def main():
    while True:
        # User input for grid
        rows = int(input("Enter number of rows: "))
        cols = int(input("Enter number of columns: "))
        grid = np.zeros((rows, cols), dtype=int)

        # User input for obstacles
        num_obs = int(input("Enter number of obstacles: "))
        print("Enter obstacle positions as row,column (0-indexed):")
        for _ in range(num_obs):
            r, c = map(int, input().split(","))
            grid[r, c] = 1

        # Display initial grid using Pandas
        print("\nInitial Grid:")
        print(pd.DataFrame(grid))

        # Start and end points
        start = tuple(map(int, input("Enter starting point row,col: ").split(",")))
        end = tuple(map(int, input("Enter destination point row,col: ").split(",")))

        if grid[start] == 1 or grid[end] == 1:
            print("Start or destination is on an obstacle! Try again.")
            continue

        # Find path using A*
        path = astar(grid, start, end)

        if path:
            print("\nPath found:")
            print(path)
            # Mark path on grid for Pandas display
            grid_path = grid.copy()
            for r, c in path:
                if (r, c) != start and (r, c) != end:
                    grid_path[r, c] = 2
            print("\nGrid with Path (2 = path):")
            print(pd.DataFrame(grid_path))
            # Visualize
            visualize_grid(grid, path, start, end)
        else:
            print("No valid path found!")

        retry = input("Do you want to try again? (yes/no): ").lower()
        if retry != "yes":
            break

# Run the program
if __name__ == "__main__":
    main()
```
